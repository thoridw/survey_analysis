## Aligning response categories for variables and counting occurrences

The grouping of the data across multiple waves concerns the variables 
- gender,
- level of education,
- income quintile,
- age,
- flood experience.

To align the data across the different waves and countries, the following is done:

##### Gender
- We connect the data on gender from the first wave (US, ID) to the data from the fifth wave.
##### Level of Education
- We connect the data on gender from the first wave (US, ID) to the data from the fifth wave. 
- Different education categories are used across the different countries. To unify, we group all categories into the following four categories:
    - no high school degree. For the different countries this includes:
        - ID: primary school (SD)
        - NL: I did not complete any formal education, primary education, lower secondary education
        - UK: no formal qualifications, youth training certificate / skill seekers
        - US: no highschool
    - high school degree. For the different countries this includes:
        - ID: junior secondary school (SMP), secondary school (SMA), vocational and pre-professional senior secondary schools (SMK)
        - NL: lower or upper secondary education
        - UK: recognized trade apprenticeship, clerical and commercial, City&Guilds certificate, ONC, CSE grades 2-5, CSE grade 1, GCE 0 level, GCSE, school certificate, Scottish ordinary/lower certificate, GCE A level or Higher certificate
        - US: high school graduate, some college
    - college degree. For the different countries this includes:
        - ID: Vocational education D3 level, Level S-1 (e.g., BA, B.Sc., B.Ed.)
        - NL: post-secondary, non-tertiary education, short-cycle tertiary education (vocational training), Bachelors or equivalent level degree
        - UK: University diploma, University or CNAA first degree (e.g., BA, B.Sc., B.Ed.), City&Guilds certificate - advances, Nursing qualification, Teaching qualification, 
        - US: 2-year, 4-year college program
    - post graduate degree. For the different countries this includes:
        - ID: Level S-2/S-3 (Masters, Doctorate), professional higher education
        - NL: Masters or equivalent level degree, Doctoral or equivalent level degree
        - UK: University or CNAA higher degree (e.g., MSc, PhD), other technical, professional or higher qualification
        - US: post-grad
#### Income quintile
- We connect the data on gender from the first wave (US, ID) to the data from the fifth wave.
- The assumption is that respondents stayed in their income quintile between the first and fifth wave (concerning US, ID).
- For ID, respondents gave their income in IDR. We translated this into income quintiles using data from the worldbank (https://pip.worldbank.org/country-profiles/IDN):
    - lowest quintile: < 19.4 million IDR
    - second-lowest quintile: 19.4-26.6 million IDR
    - middle quintile: 26.6-35.8 million IDR
    - second highest quintile: 35.8-51.1 million IDR
    - highest quintile: > 51.1 million IDR
#### Age
- We connect the data on gender from the first wave (US, ID) to the data from the fifth wave.
- The assumption is that respondents stayed in their age group between the first and fifth wave (concerning US, ID).
- Where the age was given, they were grouped into the following categories:
    - 1: 16-24 years
    - 2: 25-34 years
    - 3: 35-44 years
    - 4: 45-54 years
    - 5: 55-64 years
    - 6: 65+ years
#### Flood experiences
- For the US and ID data, we check if they answered yes to experiencing flooding in any of the waves 1-5. We also want to get the year from the last flood experience. In W1, this is question Q18b_flood_year. In wave 2-4 it is R013_recent_flood_exp. If yes in
    - w2: apr-nov2020 --> 2020
    - w3: nov2020-may2021 --> 2021
    - w4: may-nov2021 --> 2021
    - w5: Nov2021 - Jul2023 --> 2022
